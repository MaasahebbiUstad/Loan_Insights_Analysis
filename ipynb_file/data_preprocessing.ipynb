```python
# Importing necessary libraries: NumPy for numerical operations, and Pandas for data manipulation and analysis
import numpy as np
import pandas as pd
```

```python
# Reading data from an Excel file named 'saarthi_assignment.xlsx' located at '/content/' and storing it in a DataFrame named 'df'
df=pd.read_excel('/content/saarthi_assignment.xlsx')

# Displaying the first few rows of the DataFrame 'df' to inspect the imported data
df.head()
```

##Data Wrangling

**data cleaning, preprocessing, and exploratory data analysis**

```python
# Counting the number of duplicated rows in the DataFrame 'df'

df.duplicated().sum()
```

```python
# Removing duplicate rows from the DataFrame 'df' and updating it with the modified data

df=df.drop_duplicates()
```

```python
# Displaying the first row of the DataFrame 'df'

df.head(1)
```

```python
# Converting the 'loan_id' column to a string, sorting unique values in descending order

df['loan_id'].astype(str).sort_values(ascending=False).unique()
```

```python
# Counting and sorting the occurrences of unique values in the 'loan_id' column in descending order

df['loan_id'].value_counts().sort_values(ascending=False)
```

```python
# Filtering rows in the DataFrame 'df' where 'loan_id' is not equal to "i123"

df=df[df['loan_id']!="i123"]
len(df)
```

```python
# Converting the 'loan_id' column to numeric values in the DataFrame 'df'

df['loan_id']=pd.to_numeric(df['loan_id'])
```

```python
# Generating descriptive statistics for the DataFrame 'df'

df.describe()
```

```python
# Converting the 'dialed_time' column to datetime format in the DataFrame 'df'

df['dialed_time']=pd.to_datetime(df['dialed_time'])
```

```python
# Displaying the data types of each column in the DataFrame 'df'

df.dtypes
```

```python
# Displaying the last 30 rows where the 'ptp_date' column has a value of "-"

df[df['ptp_date']=="-"].tail(30)
```

```python
# Filling missing values in the 'gender' column with the value "Unknown" in the DataFrame 'df'

df['gender']=df['gender'].fillna("Unknown")
```

```python
# Counting the occurrences of unique values in the 'gender' column of the DataFrame 'df'

df['gender'].value_counts()
```

```python
# Removing time zone information from the 'emiDate' column in the DataFrame 'df'

df['emiDate'] = df['emiDate'].dt.tz_localize(None)
```

```python
# Saving the cleaned DataFrame 'df' to an Excel file named 'Saarthi_Cleaned.xlsx' without including the index

df.to_excel('Saarthi_Cleaned.xlsx', index=False)
```
